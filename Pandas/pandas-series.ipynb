### Pandas Series

A Pandas Series is like a column in a table. It is a 1-D array holding data of any type.

### Importing Pandas

```python
import numpy as np
import pandas as pd
```

```python
country = ["Pakistan", "Palestine", "Saudi Arabia", "Qatar","UAE" ]
pd.Series(country)
```

```python
# custom index
marks = [75, 59, 80, 87]
subjects = ['History', 'Science', 'Maths', 'Urdu']

pd.Series(marks, index=subjects)
```

```python
pd.Series(marks, index=subjects, name='marks of Ali')
```

### Series from dict


```python
marks = {
    'History':    75,
    'Science':    59,
    'Maths':      80,
    'Urdu':       87,
}

marks_series = pd.Series(marks, name="Ali ke marks")
marks_series
```

### Filtering a DataFrame

```python
ipl.head(2)
```

```python
# find all the final winners
mask = ipl["MatchNumber"] == "Final"
new_df = ipl[mask]
new_df[['Season', 'WinningTeam']]
```

```python
# how many super over finishes have occured
ipl[ipl["SuperOver"] == "Y"]

```

```python
ipl.head(3)
```

```python
# how many matches has csk won in kolkata
ipl[(ipl['City'] == "Kolkata") & (ipl["WinningTeam"] == "Chennai Super Kings")].shape[0]
```

```python
# The team that wins the coin toss often wins the game. 
ipl[ipl["TossWinner"] == ipl["WinningTeam"]].shape[0]/ipl.shape[0]*100
```

```python
# movies with rating higher than 8 and votes>10000
moveis[(moveis["imdb_rating"] > 8) & (moveis["imdb_votes"] > 10000)].shape[0]
```

```python
# Action movies with rating higher than 7.5

mask_1 = moveis["genres"].str.contains("Action")
mask_2 = moveis["imdb_rating"] > 7.5


moveis[mask_1 & mask_2]
```
