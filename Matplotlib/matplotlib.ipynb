```python
# from a pandas dataframe
batsman = pd.read_csv('./DataSets/sharma-kohli.csv')
batsman

plt.plot(batsman['index'], batsman['V Kohli'])
```

```python
# labels title
plt.plot(batsman['index'], batsman['V Kohli'])
plt.plot(batsman['index'], batsman['RG Sharma'])

plt.title("Rohit Sharma vs Virat Kholi IPL Comparision")
plt.xlabel("Seasons")
plt.ylabel("Score")
```

```python
# marker

plt.plot(batsman['index'], batsman['V Kohli'], color='blue', linestyle='solid', linewidth=3, marker='D', markersize = 10)
plt.plot(batsman['index'], batsman['RG Sharma'], color='#ff0000', linestyle='dashed', linewidth=2, marker='D',  markersize = 8)

plt.title("Rohit Sharma vs Virat Kholi IPL Comparision")
plt.xlabel("Seasons")
plt.ylabel("Score")
```

```python
# legend -> location

plt.plot(batsman['index'], batsman['V Kohli'], color='blue', linestyle='solid', linewidth=3, marker='D', markersize = 10, label="Virat")
plt.plot(batsman['index'], batsman['RG Sharma'], color='#ff0000', linestyle='dashed', linewidth=2, marker='D',  markersize = 8, label="Rohit")

plt.title("Rohit Sharma vs Virat Kholi IPL Comparision")
plt.xlabel("Seasons")
plt.ylabel("Score")
plt.legend()
```

```python
#  grid

plt.plot(batsman['index'], batsman['V Kohli'], color='blue', linestyle='solid', linewidth=3, marker='D', markersize = 10, label="Virat")
plt.plot(batsman['index'], batsman['RG Sharma'], color='#ff0000', linestyle='dashed', linewidth=2, marker='D',  markersize = 8, label="Rohit")

plt.title("Rohit Sharma vs Virat Kholi IPL Comparision")
plt.xlabel("Seasons")
plt.ylabel("Score")
plt.legend()
plt.grid()
```

```python
#  show

plt.plot(batsman['index'], batsman['V Kohli'], color='blue', linestyle='solid', linewidth=3, marker='D', markersize = 10, label="Virat")
plt.plot(batsman['index'], batsman['RG Sharma'], color='#ff0000', linestyle='dashed', linewidth=2, marker='D',  markersize = 8, label="Rohit")

plt.title("Rohit Sharma vs Virat Kholi IPL Comparision")
plt.xlabel("Seasons")
plt.ylabel("Score")
plt.legend()
plt.grid()
plt.show()
```

# Scatter Plots

image.png

- Bivariate Analysis
- numerical vs numerical
- Use case - Finding correlation

```python
# plt.scatter simple function

total_bill = [15, 20, 25, 30, 35, 40, 45, 50, 60, 70]
tip_amount = [2.00, 3.50, 4.00, 5.00, 7.00, 8.50, 10.00, 12.00, 15.00, 18.00]


plt.scatter(total_bill, tip_amount, color='blue', marker='o')

plt.title('Tips vs. Total Bill Amount')
plt.xlabel('Total Bill ($)')
plt.ylabel('Tip Amount ($)')
# plt.grid()
```

## Bar chart

image.png

- Bivariate Analysis
- Numerical vs Categorical
- Use case - Aggregate analysis of groups

```python
# simple bar chart

products = ['Apples', 'Bananas', 'Oranges', 'Grapes', 'Strawberries', 
            'Pineapples', 'Watermelons', 'Blueberries', 'Kiwis', 'Mangoes']
sales = [120, 150, 100, 180, 160, 90, 140, 110, 80, 130]

plt.figure(figsize=(12, 6))

plt.bar(products, sales,color='yellow')

plt.xlabel('Products')
plt.ylabel('Sales ($)')
```

```python
# horizontal bar chart 

plt.barh(products, sales,color='yellow')

```

```python
# color and label
df = pd.read_csv('./DataSets/batsman_season_record.csv')
df
```

```python
plt.bar(np.arange(df.shape[0]) - 0.2,df['2015'],width=0.2 )
plt.bar(np.arange(df.shape[0]),df['2016'],width=0.2 )
plt.bar(np.arange(df.shape[0]) + 0.2,df['2017'],width=0.2 )

plt.xticks(np.arange(df.shape[0]), df['batsman'])

plt.show()
```

```python
# Stacked Bar chart
plt.bar(df['batsman'],df['2017'],label='2017')
plt.bar(df['batsman'],df['2016'],bottom=df['2017'],label='2016')
plt.bar(df['batsman'],df['2015'],bottom=(df['2016'] + df['2017']),label='2015')

plt.legend()
plt.show()
```

## Pie Chart

image.png

- Univariate/Bivariate Analysis
- Categorical vs numerical
- Use case - To find contibution on a standard scale

```python
# Simple piechart
categories = ['Fruits', 'Vegetables', 'Dairy', 'Grains', 'Proteins']
values = [30, 25, 20, 15, 10]


plt.pie(values, labels=categories)
plt.show()
```

```python
# On dataset

df = pd.read_csv('./DataSets/gayle-175.csv')
df
```

```python
plt.pie(df['batsman_runs'], labels=df['batsman'], autopct='%0.1f%%')
plt.show()
```

```python
# percentage and colors
plt.pie(df['batsman_runs'], labels=df['batsman'], autopct='%0.1f%%', colors=['blue', 'yellow', 'black', 'brown', 'pink','cyan'])
plt.show()
```

```python
# explode, shadow
plt.pie(df['batsman_runs'],labels=df['batsman'],autopct='%0.1f%%',explode=[0.3,0,0,0,0,0.1],shadow=True)
plt.show()
```
